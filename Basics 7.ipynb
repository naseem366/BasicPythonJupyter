```python
#Display only those lines which are starting with 'how'
import re

hand = open('sample.txt')

for line in hand:
    line=line.rstrip()
    if re.search('^how',line): # ^ for the starting
        print(line)
        
hand.close()
```

## Exercises:
(use input file: mbox-short.txt)
1. Write a program to extract lines starting with the word 'From' (or 'from') and ending with 'edu'
2. Extract all the lines ending with any digit
3. Extract the lines starting with anything other than lower-case alphabets and digits
4. The line should start with capital letters, followed by zero or more characters, but must end with any digit.

```python
#* zero 
#^ starting 
# $ it means ending 
file=open('mbox-short.txt')
for line in file:
    line=line.strip()
    if re.search('^[Ff]rom.*edu$',line):
        print(line)
file.close()
```

```python
#1
fhand = open('mbox-short.txt')
for line in fhand:
    line = line.strip()
    if re.search('^[Ff]rom.*edu$',line):
        print(line)
        
fhand.close()
```

```python
#2 #here  using /d  which means having digit [0-9 ] return any value
fhand = open('mbox-short.txt')
for line in fhand:
    line = line.strip()
    if re.search('\d$',line):
        print(line)
        
fhand.close()
```

```python
#3
fhand = open('mbox-short.txt')
for line in fhand:
    line = line.strip()
    if re.search('^[^a-z0-9]',line):
        print(line)
        
fhand.close()
```

```python
#4
fhand = open('mbox-short.txt')
for line in fhand:
    line = line.strip()
    if re.search('^[A-Z].*\d$',line):
        print(line)
        
fhand.close()
```

5. Write a program to extract all the email IDs from the file mbox-short.txt

```python
import re
fhand = open('mbox-short.txt')
ls=[]
for line in fhand:
    line = line.strip()
    lst= re.findall('[a-zA-Z0-9]\S+@\S+[a-zA-z][.][a-zA-Z]+',line)
    if(len(lst))>0:
        ls.extend(lst)
        
print(ls)
    
        
fhand.close()
```

```python
fhand = open('mbox-short.txt')
ls1=list()
for line in fhand:
    line = line.strip()
    lst= re.findall('\S+@\S+',line)
    if(len(lst))>0:
        ls1.extend(lst)
#print(ls1)

for i in range(len(ls1)):
    ls1[i]=ls1[i].strip('<>;)')
    
print(ls1)    
fhand.close()
```
